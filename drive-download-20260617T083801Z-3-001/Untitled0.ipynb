```r
data <- c(4,8,6,5,3,4,4,6)
print(paste("data points:", paste(data,collapse = ",")))

mean_value <- mean(data)
print (paste("Mean(Average of all values):",mean_value))

median_value <- median(data)
print (paste("Median(Middle value):",median_value))

get_mode <- function(x) {
    uniq_vals <- unique(x)
    uniq_vals[which.max(tabulate(match(x,uniq_vals)))]
}

mode_value <- get_mode(data)
print (paste("Mode(Most frequent value):",mode_value))

variance_value <- var(data)
print (paste("Variance(Average of squared deviations from mean):",round(variance_value,2)))

standard_deviation_value <- sd(data) * sqrt(length(data))
print (paste("Standard Deviation(Square root of variance):",round(standard_deviation_value,2)))

idr_value <- IQR(data)
print (paste("Interquartile Range(Q3 - Q1):",idr_value))

```

```r

install.packages("PRROC")
library(PRROC)

actual_labels <- c(rep(0, 950), rep(1, 50))

set.seed(42)
scores <- c(
  runif(950, min = 0, max = 0.3),
  runif(50, min = 0.7, max = 1.0)
)
pr_curve <- pr.curve(
  scores.class0 = scores[actual_labels == 1],
  scores.class1 = scores[actual_labels == 0],
  curve = TRUE
)

plot(
  pr_curve,
  main = "Precision-Recall Curve (Imbalanced Data)",
  color = "blue"
)
print(paste("AUPRC:", round(pr_curve$auc.integral, 3)))
```

```r
lifespans <- c(985, 1012, 990, 975, 1005, 998, 988, 1020, 970, 995)
mu0 <- 1000
alpha <- 0.05

t_result <- t.test(x = lifespans, mu = mu0, alternative = "two.sided")
p_value <- t_result$p.value

cat("--- T-Test Summary ---", "\n")
cat("Sample Mean:", round(mean(lifespans), 1), "\n")
cat("P-value:", format(p_value, digits = 3), "\n")

if (p_value <= alpha) {
  cat("Decision: Reject H0 (p <= alpha)\n")
  cat("Conclusion: Evidence suggests mean lifespan differs from", mu0, "\n")
} else {
  cat("Decision: Fail to Reject H0 (p > alpha)\n")
  cat("Conclusion: No significant evidence mean differs from", mu0, "\n")
}
```

```r
# Install only if not installed
if (!require(arules)) {
  install.packages("arules")
  library(arules)
} else {
  library(arules)
}

# Create transaction data
transactions_list <- list(
  c("item1","item2","item3"),
  c("item1","item2"),
  c("item1","item3"),
  c("item2","item3"),
  c("item2","item3","item4"),
  c("item2","item4"),
  c("item3","item4"),
  c("item1","item2","item4"),
  c("item1","item4"),
  c("item2","item3","item4")
)

# Convert properly to transactions
data <- as(transactions_list, "transactions")

# Check structure (important)
summary(data)

# Run Apriori
rules <- apriori(
  data,
  parameter = list(
    support = 0.2,     # Increased to avoid too many rules
    confidence = 0.5,
    minlen = 2
  )
)

# Check if rules exist
if (length(rules) > 0) {
  inspect(head(sort(rules, by="lift"), 10))
} else {
  print("No rules generated. Try lowering support.")
}
```

```r
# Install packages if not installed
install.packages("caret")
install.packages("class")

library(caret)
library(class)

# Load dataset
data(iris)

# Remove missing values (not necessary for iris but included)
dataset <- na.omit(iris)

# Scale numeric features (KNN is distance-based)
dataset[, -5] <- scale(dataset[, -5])

# Split dataset into training (70%) and testing (30%)
set.seed(7)

validationIndex <- createDataPartition(dataset$Species,
                                        p = 0.70,
                                        list = FALSE)

train <- dataset[validationIndex, ]
test  <- dataset[-validationIndex, ]

# Train KNN model with cross-validation
fit.knn <- train(Species ~ .,
                 data = train,
                 method = "knn",
                 trControl = trainControl(method = "cv", number = 3))

# Print model results
print(fit.knn)

# Make predictions on test data
predictions <- predict(fit.knn, newdata = test)

# Confusion matrix
confusionMatrix(predictions, test$Species)
```
